<b>w/ subplot, you can plot 2 types of graphs, line and bar</b>

```python
import matplotlib.pyplot as plt

fig = plt.figure()
ax1 = fig.add_subplot(2, 1, 1)
ax2 = fig.add_subplot(2, 1, 2)

x = range(0, 100)
y = [v*v for v in x]

ax1.plot(x, y)
ax2.bar(x, y)

plt.show()
```

### 3. Label and Legend

<br><br>

<b> Labeling a graph is easy, use set_xlabel and set_ylabel method</b>

```python
sin_y = np.sin(x)
cos_y = np.cos(x)

fig = plt.figure()
ax1 = fig.add_subplot(2, 1, 1)
ax2 = fig.add_subplot(2, 1, 2)

ax1.plot(x, sin_y, 'b--')
ax2.plot(x, cos_y, 'r--')
```

```python
ax1.set_xlabel('x')
ax1.set_ylabel('sin(x)')

ax2.set_xlabel('x')
ax2.set_ylabel('cos(x)')
```

```python
plt.show()
```

<br><br>
<b>Let's plot two graph in the same subplot, and add a lengend</b>

```python
sin_y = np.sin(x)
cos_y = np.cos(x)

plt.plot(x, sin_y, label='sin(x)')
plt.plot(x, cos_y, label='cos(x)')

plt.legend(loc='upper left')
```

```python
plt.show()
```

### 5. Various Types of Graph

<br><br>

<b>This time we will map adj close price and trading volume in the same window</b>

```python
import matplotlib.pyplot as plt
import pandas_datareader.data as web
```

```python
sk_hynix = web.DataReader("000660.KS", "yahoo")
```

```python
fig = plt.figure(figsize=(12, 8))
```

```python
fig
```

```python
plt.show()
```

```python
top_axes = plt.subplot2grid((4,4), (0,0), rowspan=3, colspan=4)
bottom_axes = plt.subplot2grid((4,4), (3,0), rowspan=1, colspan=4)
```

```python
plt.show()
```

<br><br>
<b>This time we used subplot2grid method  <br><br>
first argument (4,4) meaning dividing the whole figure in 4x4 grid,  
(0,0) row and colspan meaning starting from upperleft(0,0),  
making a AxesSubplot of 4 columns wide and 3 rows height</b>

<br><br>
<b>same for bottom axes, dividing the whole figure in 4x4 grid.  
starting from (3,0) which is bottom left,  
making a AxesSubplot of 4 columns wide and 1 row height</b>
<br><br>

<b>Code below prevents the trading volume presenting as the scientific notation</b>

```python
bottom_axes.get_yaxis().get_major_formatter().set_scientific(False)
```

```python
top_axes.plot(sk_hynix.index, sk_hynix['Adj Close'], label='Adjusted Close')
```

```python
bottom_axes.plot(sk_hynix.index, sk_hynix['Volume'])
```

<br><br>
<b>Code below makes the layout more tight, making subplots as big as possible</b>

```python
plt.tight_layout()
```

```python
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
from matplotlib import font_manager, rc
from matplotlib import style
```

```python
font_name = font_manager.FontProperties(fname="C:/Users/이승민/AppData/Local/Microsoft/Windows/Fonts/NotoSansKR-Medium.otf").get_name()
rc('font', family=font_name)
```

```python
style.use('ggplot')
```

```python
industry = ['통신업', '의료정밀', '운수창고업', '의약품', '음식료품', '전기가스업', '서비스업', '전기전자', '종이목재', '증권']
fluctuations = [1.83, 1.30, 1.30, 1.26, 1.06, 0.93, 0.77, 0.68, 0.65, 0.61]
```

```python
fig = plt.figure(figsize=(12, 8))
ax = fig.add_subplot(111)
```

```python
ypos = np.arange(10)
rects = plt.barh(ypos, fluctuations, align='center', height=0.5)
plt.yticks(ypos, industry)
```

```python
for i, rect in enumerate(rects):
    ax.text(0.95 * rect.get_width(), rect.get_y() + rect.get_height() / 2.0, str(fluctuations[i]) + '%', ha='right', va='center')
```

```python
plt.xlabel('등락률')
```

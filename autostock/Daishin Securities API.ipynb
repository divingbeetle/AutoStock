### 4. Past Data

<br><br>

<b>StockChart Class from CpSysDib Module provides past data of a stock</b>

<b>Let's create a instance first</b>

```python
import win32com.client
instStockChart = win32com.client.Dispatch("CpSysDib.StockChart")
```

<b>StockChart Class uses Request/Reply Protocol  
Use SetInputValue method to request</b>

<b>object.SetInputValue(type, value)  
Type: Value
+ 0: StockCode(String)   
    
+ 1: Request type(char)  
    '1' for period
    '2' for num
+ 2: Request EndTime(ulong)      
    YYYYMMDD, Default(0) Recent exchange
+ 3: Request StartTime(ulong)   
    YYYYMMDD
+ 4: Request number(ulong)    
    num of request data
+ 5: Field(long array)     
    date, time, price, ... 
+ 6: Chart Type(Char)     
    'D' for day
    'T' for tick ...
  
    for detailed info, check [here](https://money2.daishin.com/e5/mboard/ptype_basic/HTS_Plus_Helper/DW_Basic_Read_Page.aspx?boardseq=284&seq=102&page=1&searchString=CpSysDib.StockChart&p=&v=&m=)
</b>

```python
instStockChart.SetInputValue(0, "A003540")
instStockChart.SetInputValue(1, ord('2'))
instStockChart.SetInputValue(4, 10)
instStockChart.SetInputValue(5, 5)
instStockChart.SetInputValue(6, ord('D'))
instStockChart.SetInputValue(9, ord('1'))
```

<br><br>
<b>Code above implies
+ 0: data about "A003540" ('대신증권')
+ 1: number of data
+ 4: 10 number of data
+ 5: 10 number of close price
+ 6: 10 number of close price of the day
+ 9: 10 numver of 'adjusted' close price of the day  
</b>

<br><br>
<b>After initial setting, we can request the data </b>

```python
instStockChart.BlockRequest()
```

<b>when you request a data, server reploes  
Use GetHeaderValue and GetDataValue method to get data</b>

```python
numData = instStockChart.GetHeaderValue(3)
```

```python
numData
```

<b>number of data is valid, now print the close price of the 10 days</b>

```python
for i in range(numData):
    print(instStockChart.GetDataValue(0, i))
```

<b>We only requested 1 field, close price of the stock.

### 6. Setting Algorithm

<br><br>
<b>Let's set up a simple trading algorithm for excercise</b>

<b>Our first alorithm will find  
+ 1) Massive Trading Volume : increasing more than 1,000%
+ 2) Low PBR : PBR \<4 at Massive Trading </b>

<br><br>
<b>To get the Trading Volume, we can use StockChart Class  
below is a code saving recent 60 days' trading volume of certain stock </b>

```python
import win32com.client
instStockChart = win32com.client.Dispatch("CpSysDib.StockChart")
```

```python
instStockChart.SetInputValue(0, "A003540")
instStockChart.SetInputValue(1, ord('2'))
instStockChart.SetInputValue(4, 60)
instStockChart.SetInputValue(5, 8)
instStockChart.SetInputValue(6, ord('D'))
instStockChart.SetInputValue(9, ord('1'))
```

```python
instStockChart.BlockRequest()
```

```python
numdata = instStockChart.GetHeaderValue(3)
volumes = [instStockChart.GetDataValue(0, i) for i in range(numdata)]
```

```python
volumes
```

<br><br>
<b>Now we wil compare recent trading volume w/ 59 days' average  
volumes[0] will ve our recent trading volume</b>

```python
averageVolume = (sum(volumes) - volumes[0]) / (len(volumes) - 1)
```

```python
if volumes[0] > averageVolume * 10:
    print("Condition: True")
else:
    print("Condition: False, ", volumes[0] / averageVolume)
```

### 7. Comparing PER in the Sector

<br><br>
<b>PER is a great index for investment.  
However, PER can differ from industry sector or company size.</b>

<br><br>
<b>Let's compare the company's PER w/ its industry's average   
by using GetIndustryList method</b>

```python
import win32com.client
```

```python
instCpCodeMgr = win32com.client.Dispatch("CpUtil.CpCodeMgr")
```

```python
industryCodeList = instCpCodeMgr.GetIndustryList()
```

```python
for industryCode in industryCodeList:
    print(industryCode, instCpCodeMgr.GetIndustryName(industryCode))
```

<br><br>
<b>Let's say we are interested in the Food Industry</b>

```python
instCpCodeMgr = win32com.client.Dispatch("CpUtil.CpCodeMgr")
tarketCodeList = instCpCodeMgr.GetGroupCodeList(5) # 5 for '음식료품'

for code in tarketCodeList:
    print(code, instCpCodeMgr.CodeToName(code))
```

<br><br>
<b>Let's load average PER of the food industry, w/ MarketEye Class</b>

```python
instMarketEye = win32com.client.Dispatch("CpSysDib.MarketEye")
```

```python
# Get PER
instMarketEye.SetInputValue(0, 67)
instMarketEye.SetInputValue(1, tarketCodeList)
```

```python
instMarketEye.BlockRequest()
```

```python
numStock = instMarketEye.GetHeaderValue(2)
```

```python
sumPER = sum((instMarketEye.GetDataValue(0, i) for i in range(numStock)))
```

```python
print("Average PER of Food Industry: ", sumPER / numStock)
```
